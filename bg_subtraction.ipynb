```python
#averaging-background subtraction
cap = cv2.VideoCapture(r"C:\Users\hanal\Downloads\13364107_2160_3840_60fps.mp4")
background = None
while True:
    r, f = cap.read()
    if not r:
        break
    g = cv2.cvtColor(f, cv2.COLOR_BGR2GRAY)
    gf=cv2.resize(f,(500,500))
    if background is None:
        background = gf.astype("float")
        continue
    cv2.accumulateWeighted(gf, background, 0.01) 
    diff_frame = cv2.absdiff(gf, cv2.convertScaleAbs(background))
    _, fg_mask = cv2.threshold(diff_frame, 25, 255, cv2.THRESH_BINARY)
    cv2.imshow('original',gf)
    cv2.imshow('Foreground Mask', fg_mask)
    if cv2.waitKey(30) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

```python
#scene change using accumulated weight
import numpy as np
def detect(video_path, threshold=50000, learning_rate=0.01):
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print("Error: Unable to open video.")
        return
    background = None
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        gray_frame=cv2.resize( gray_frame,(500,500))
        if background is None:
            background = gray_frame.astype("float")
            continue
        cv2.accumulateWeighted(gray_frame, background, learning_rate)
        diff_frame = cv2.absdiff(gray_frame, cv2.convertScaleAbs(background))
        _, fg_mask = cv2.threshold(diff_frame, 25, 255, cv2.THRESH_BINARY)
        changed_pixels = cv2.countNonZero(fg_mask)
        if changed_pixels > threshold:
            print("Scene change detected!")
        cv2.imshow('original',gray_frame)
        cv2.imshow('Foreground Mask', fg_mask)
        k=cv2.waitKey(30) 
        if k == 30:
            break
    cap.release()
    cv2.destroyAllWindows()
    
detect(r"C:\Users\hanal\Downloads\8394092-uhd_2160_3840_24fps.mp4")
```

```python
#motion detection with scene change
c = cv2.VideoCapture(r"C:\Users\hanal\Downloads\8394092-uhd_2160_3840_24fps.mp4")
bg = cv2.createBackgroundSubtractorMOG2(history=500, varThreshold=100, detectShadows=False)
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))
ret, preframe = c.read()
if not ret:
    print("Error: Could not read first frame")
    c.release()
    exit()
preframe = cv2.resize(preframe, (500, 500))  
preframe = cv2.cvtColor(preframe, cv2.COLOR_BGR2GRAY)
while True:
    ret, frame = c.read()
    if not ret:
        break
    frame_resized = cv2.resize(frame, (500, 500)) 
    gray = cv2.cvtColor(frame_resized, cv2.COLOR_BGR2GRAY)
    diff = cv2.absdiff(preframe, gray)
    _, diff_thresh = cv2.threshold(diff, 50, 255, cv2.THRESH_BINARY)
    if np.sum(diff_thresh) > 5_000_000:  
        print("Scene Change Detected: Resetting Background Model")
        bg = cv2.createBackgroundSubtractorMOG2(history=500, varThreshold=100, detectShadows=False)
    preframe = gray.copy() 
    fg = bg.apply(frame_resized)
    fg = cv2.morphologyEx(fg, cv2.MORPH_OPEN, kernel)
    fg = cv2.morphologyEx(fg, cv2.MORPH_CLOSE, kernel)
    cv2.imshow('Original', frame_resized)
    cv2.imshow('Motion Detection', fg)
    if cv2.waitKey(30) & 0xFF == 27: 
        break
c.release()
cv2.destroyAllWindows()
```
