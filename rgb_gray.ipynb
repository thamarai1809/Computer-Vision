### Reading the image


```python
import matplotlib.pyplot as plt
import cv2
```

```python
img=cv2.imread('lenna.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

### RGB to Grayscale 

```python
import cv2
import matplotlib.pyplot as plt

# Assuming img is your original image in RGB format
gray_image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

# Display the grayscale image correctly
plt.imshow(gray_image, cmap='gray')

plt.show()
```

### Without inbuilt function:
#### 1.Normal average
#### 2.Weighted average

### Normal average

```python
import matplotlib.pyplot as plt

(row, col) = img.shape[0:2]

for i in range(row):
    for j in range(col):
        img[i, j] = sum(img[i, j]) * 0.33

plt.imshow(img)
plt.title('Gray Scale image')

```

### Weighted average

```python
image_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

r = image_rgb[:,:,0]
g = image_rgb[:,:,1]
b = image_rgb[:,:,2]


grayscale_image = 0.299 * r + 0.587 * g + 0.114 * b

plt.imshow(grayscale_image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image_path = 'lenna.jpg'  
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)  

# Apply Histogram Equalization
equalized_image = cv2.equalizeHist(image)

# Plot original and equalized images side by side
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Original Image
axs[0].imshow(image, cmap='gray')
axs[0].set_title('Original Image')
axs[0].axis('off')

# Equalized Image
axs[1].imshow(equalized_image, cmap='gray')
axs[1].set_title('Equalized Image')
axs[1].axis('off')


fig, axs = plt.subplots(1, 2, figsize=(12, 6))


axs[0].hist(image.ravel(), bins=256, color='red', alpha=0.7)
axs[0].set_title('Histogram of Original Image')

# Histogram for Equalized Image
axs[1].hist(equalized_image.ravel(), bins=256, color='red', alpha=0.7)
axs[1].set_title('Histogram of Equalized Image')

plt.show()
```
